```python
zips = pd.read_csv("clean_operational_ny_zips.csv")
```

```python
zips.rename({"Zip Code": "Zipcode"}, axis=1, inplace=True)
```

```python
df = pd.merge(left=census, right=zips, on="Zipcode", how="right")
```

```python
restaurant_counts = {}
```

```python
for i, zip_code in enumerate(df.Zipcode):
    if zip_code in restaurant_counts:
        continue
    PARAMETERS = {'term': 'restaurant',#An example
              'location': '{}'.format(zip_code)}
    
    response = requests.get(url = ENDPOINT,
                        params = PARAMETERS,
                        headers = HEADERS)
    data = response.json()
    try:
        restaurant_counts[zip_code] = data['total']
        print(zip_code, data['total'])
        if (i + 1) % 50 == 0:
            print("%5d / %5d"%(i + 1, len(census)))
    except Exception as e:
        print("Error {} {}".format(zip_code, str(e)))
```
